```python
import numpy as np
import matplotlib.pyplot as plt
import pycascades as pc
import networkx as nx
import seaborn as sns
sns.set(font_scale=1.5)
sns.set_style("white")
```

We first create a cusp and a linear coupling.

```python
cusp_element_0 = pc.cusp( a = -4, b = 1, c = 0, x_0 = 0.5 )
coupling_0 = pc.linear_coupling( strength = 0.05 )
```

We create an Erdös-Rényi random graph.

```python
size = 15
degree = 3
rewiring_ws = 0.15

G = nx.erdos_renyi_graph(size, degree/size, directed = True, seed = None)
net = pc.gen.networks.from_nxgraph( G, element_pool = [cusp_element_0], coupling_pool = [coupling_0])

print(f"Average degree of network {net.number_of_edges() / net.number_of_nodes()}")
while net.number_of_edges() / net.number_of_nodes() < 2.9 or net.number_of_edges() / net.number_of_nodes() > 3.1:
	G = nx.erdos_renyi_graph(size, degree/size, directed = True, seed = None)
	net = pc.gen.networks.from_nxgraph( G, element_pool = [cusp_element_0], coupling_pool = [coupling_0])

plt.xticks([])
plt.yticks([])
#plotter.network(net).savefig('figures/er.png')
#plotter.network(net).savefig('figures/er.pdf')
pc.plotter.network(net).show()
#Average Degree
av_degree = net.number_of_edges() / net.number_of_nodes()
print(f"Average degree of network {net.number_of_edges() / net.number_of_nodes()}")
```

Watts-Strogatz network

```python
net = pc.gen.networks.directed_watts_strogatz_graph(size, degree, rewiring_ws, element_pool = [cusp_element_0], coupling_pool = [coupling_0] )
#Matplotlib Specifications
plt.xticks([])
plt.yticks([])
#plotter.network(net).savefig('figures/ws.png')
#plotter.network(net).savefig('figures/ws.pdf')
pc.plotter.network(net).show()
#Average Degree
av_degree = net.number_of_edges() / net.number_of_nodes()
print(f"Average degree of network {av_degree}")
```

Barabasi-Albert (scale-free) network

```python
net = pc.gen.networks.directed_barabasi_albert_graph(size, degree, element_pool = [cusp_element_0], coupling_pool = [coupling_0] )
#Matplotlib Specifications
plt.xticks([])
plt.yticks([])
#plotter.network(net).savefig('figures/ba.png')
#plotter.network(net).savefig('figures/ba.pdf')
pc.plotter.network(net).show()
#Average degree
av_degree = net.number_of_edges() / net.number_of_nodes()
print(f"Average degree of network {av_degree}")
```
